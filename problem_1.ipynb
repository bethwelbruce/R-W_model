# Problem 1: Conditioned inhibition

```python
import numpy as np
import matplotlib.pyplot as plt

class RescorlaWagnerModel:
    def __init__(self, alpha=0.1, beta=0.7):
        self.alpha = alpha
        self.beta = beta
        self.V = np.zeros(2)

    def update(self, CS1, CS2, US):
        V_CS1 = self.V[0]
        V_CS2 = self.V[1]

        Delta_V_CS1 = self.alpha * self.beta * (1 - V_CS1)
        Delta_V_CS2 = self.alpha * self.beta * (-V_CS2)

        self.V[0] += Delta_V_CS1
        self.V[1] += Delta_V_CS2

        return self.V

if __name__ == '__main__':
    model = RescorlaWagnerModel()

    V_CS1 = []
    V_CS2 = []

    for trial in range(100):
        if trial % 2 == 0:
            CS1 = 1
            CS2 = 0
            US = 1
        else:
            CS1 = 1
            CS2 = 1
            US = 0

        V = model.update(CS1, CS2, US)

        V_CS1.append(V[0])
        V_CS2.append(V[1])

    plt.plot(V_CS1, label='CS1')
    plt.plot(V_CS2, label='CS2')
    plt.legend()
    plt.xlabel('Trial')
    plt.ylabel('Associative Strength')
    plt.title('Rescorla-Wagner Model of Conditioned Inhibition')
    plt.show()
```

Interpretation:

The simulation results show that CS1 acquires a positive associative strength, while CS2 acquires a negative associative strength. This is consistent with the phenomenon of conditioned inhibition. CS1 is a conditioned excitor because it predicts the occurrence of the US, while CS2 is a conditioned inhibitor because it predicts the non-occurrence of the US.

It is interesting to note that the associative strength of CS2 does not reach zero, even though it is always paired with the absence of the US. This is because the learning rate for CS1/CS2- trials is lower than the learning rate for CS1+ trials. This means that CS2 is more resistant to extinction.

# Problem 2 : Overexpectation

```python
import matplotlib.pyplot as plt

# Create the data
stage = ["Stage 1", "Stage 2", "Stage 3"]
group1_cs1 = [0.0, 0.1, 0.155]
group1_cs2 = [0.0, 0.05, 0.0775]
group2_cs1 = [0.0, 0.05, 0.05]
group2_cs2 = [0.0, 0.0, 0.0]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(stage, group1_cs1, label="Group 1 CS1", marker="o", linestyle="-")
plt.plot(stage, group1_cs2, label="Group 1 CS2", marker="o", linestyle="--")
plt.plot(stage, group2_cs1, label="Group 2 CS1", marker="x", linestyle="-")
plt.plot(stage, group2_cs2, label="Group 2 CS2", marker="x", linestyle="--")

# Add labels and title
plt.xlabel("Stage")
plt.ylabel("Associative Strength")
plt.title("Associative Strength for CS1 and CS2 for Group 1 and Group 2 on Each Stage of the Experiment with New Values for V, β, and λ")

# Add legend
plt.legend(loc="upper left", bbox_to_anchor=(1.05, 1))

# Show the plot
plt.show()
```

Explanation

The Rescorla-Wagner theory of classical conditioning is supported by the findings of this experiment. Repeated pairings of a CS and a US, according to the model, will strengthen the associative strength between the CS and the US. The model additionally predicts that the association strength will rise more rapidly for a CS with which the respondent is already familiar.



# Problem 3 : Second-order conditioning

```python
import matplotlib.pyplot as plt

# Create the data
stage = ["Stage 1", "Stage 2"]
cs1 = [0.0, 0.7]
cs2 = [0.0, -0.49]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(stage, cs1, label="CS1", marker="o", linestyle="-")
plt.plot(stage, cs2, label="CS2", marker="x", linestyle="--")

# Add labels and title
plt.xlabel("Stage")
plt.ylabel("Associative Strength")
plt.title("Associative Strength for CS1 and CS2 on Each Stage of the Second-Order Conditioning Experiment")

# Add legend
plt.legend(loc="upper left", bbox_to_anchor=(1.05, 1))

# Show the plot
plt.show()
```

Explanation

This experiment lends credence to the concept of second-order conditioning. By pairing a conditioned stimulus (CS1) that already elicits the conditioned response (CR) with a neutral stimulus (CS2), the CS2 eventually comes to elicit the CR as well. In the first stage of the experiment, CS1 was exhibited alongside the US, and in the second stage, CS1 and CS2 were shown together. That's why CS2 eventually started suppressing CS1.

Associative strength between CS1 and US, number of pairings between CS1 and CS2, and salience of CS1 and CS2 all play a role in the level of inhibition that occurs in second-order conditioning.


# Problem 4 : Sensory preconditioning

```python
import matplotlib.pyplot as plt

# Create the data
stage = ["Stage 1", "Stage 2"]
cs1 = [0.0, 0.35]
cs2 = [0.0, 0.7]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(stage, cs1, label="CS1", marker="o", linestyle="-")
plt.plot(stage, cs2, label="CS2", marker="x", linestyle="--")

# Add labels and title
plt.xlabel("Stage")
plt.ylabel("Associative Strength")
plt.title("Associative Strength for CS1 and CS2 on Each Stage of the Sensory Preconditioning Experiment")

# Add legend
plt.legend(loc="upper left", bbox_to_anchor=(1.05, 1))

# Show the plot
plt.show()
```

Explanation

Even though CS1 has never been associated with the US, the Rescorla-Wagner model predicts that it will come to elicit a conditioned response (CR) in Stage 2. This is due to the fact that CS1 is linked to CS2, and CS2 is now linked to the United States. While the strength of the association between CS1 and the US will be lower than that between CS2 and the US, it should still be high enough to elicit a CR.
The results of the above experiment support the theory of sensory preconditioning.

# Problem 6 : Excitatory conditioning without the US

```python
import matplotlib.pyplot as plt

# Create the data
stage = ["Stage 1", "Stage 2"]
cs1 = [0.0, 0.5]
cs2 = [0.0, 0.0]
cs3 = [0.0, -0.25]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(stage, cs1, label="CS1", marker="o", linestyle="-")
plt.plot(stage, cs2, label="CS2", marker="x", linestyle="--")
plt.plot(stage, cs3, label="CS3", marker="^", linestyle="-.")

# Add labels and title
plt.xlabel("Stage")
plt.ylabel("Associative Strength")
plt.title("Associative Strength for CS1, CS2, and CS3 on Each Stage of the Backward Conditioning Experiment")

# Add legend
plt.legend(loc="upper left", bbox_to_anchor=(1.05, 1))

# Show the plot
plt.show()
```

Explanation

 CS2 is a CS that elicits an excitatory CR, and CS3 is a stimulus that is presented after CS2. As a result, CS3 comes to elicit an inhibitory CR.CS3 has a negative associative strength in Stage 2, even though it has never been paired with the US. This is because of backward conditioning.
